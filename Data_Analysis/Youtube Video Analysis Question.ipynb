##### Her bir videoda kullanılan etiket sayısını hesaplayarak, oluşan değerleri yeni bir sütun olarak ekleyin.

Not: Her bir etiket "|" işaretiyle ayrılmaktadır.

```python
def numTags(x):
    tags = x.split("|")
    return len(tags)
dataset["Number of Tags"] = dataset["tags"].apply(numTags)
```

##### (Challenge) Videoları Beğeni Oranı en yüksek olandan en düşüğe doğru sıralayınız

```python
def likeRate(likes,dislikes):
    numLikes = list()
    numDislikes = list()
    for i,j in likes.iteritems():
        numLikes.append(j)
    for i,j in dislikes.iteritems():
        numDislikes.append(j)
    likeRatio = list()
    for like,dislike in zip(likes,dislikes):
        if(like+dislike==0):
            likeRatio.append(0)
        else:
            likeRatio.append(like/(like+dislike))
    return likeRatio
dataset["like_ratio"] = likeRate(dataset["likes"],dataset["dislikes"])
        
```

```python
dataset.sort_values(by = "like_ratio",ascending = False,inplace = True)
```
